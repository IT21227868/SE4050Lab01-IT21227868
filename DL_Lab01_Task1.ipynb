Create a Random 100000x1 Array and Visualize Distributions


```python
import numpy as np
import matplotlib.pyplot as plt

exp_array = np.random.exponential(scale=1.0, size=100000)
uniform_array = np.random.rand(100000)
normal_array = np.random.randn(100000)

plt.hist(exp_array, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(uniform_array, density=True, bins=100, histtype="step", color="red", label="Uniform")
plt.hist(normal_array, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

3D Plot using Matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Pokemon (2).csv', encoding='ISO-8859-1')
features = data[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Calculate Pearson and Spearman correlations
pearson_corr = features.corr(method='pearson')
spearman_corr = features.corr(method='spearman')

# Visualize the correlations using heatmaps
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.heatmap(pearson_corr, annot=True, ax=ax[0], cmap='coolwarm')
ax[0].set_title('Pearson Correlation')
sns.heatmap(spearman_corr, annot=True, ax=ax[1], cmap='coolwarm')
ax[1].set_title('Spearman Correlation')
plt.show()
```
